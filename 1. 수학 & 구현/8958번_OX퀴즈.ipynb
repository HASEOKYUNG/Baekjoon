## 8958번
O와 X로 이루어진 문자열 내 O 1개당 1점을 가질 때 연속된 O는 +1점을 갖도록 한다. 최종 점수를 구하여라.
- 입력) 첫 줄엔 점수 출력 Case 개수가 주어지고 이후 O와 X로 이루어진 문자열이 입력된다.

- 출력) 점수를 출력한다.

- https://www.acmicpc.net/problem/8958

### 방법1
- 무시해도 되는 X를 기준으로 split하면 연속되지 않은 O와 연속된 O들로 구분할 수 있다.<br>
  이후 O 단위당 O 개수에 따라 range로 점수를 합하면 된다.
- 연속된 O별로 range가 동작하기에 시간복잡도가 크다.

```python
# 30840 KB, 76 ms
for i in range(int(input())):
    print(sum([sum(range(1,len(i)+1)) for i in input().split('X') if i != '']))
```

### 방법2
- 출력점수와 O의 점수를 저장하는 두 변수를 사용해 연속된 O가 나올 경우 직전까지의 O점수에 1을 더해 출력점수에 더한다.
- X가 나올때는 O의 점수를 저장한 변수를 초기화한다.

```python
# 31256 KB, 52 ms
for i in range(int(input())):
    ox = input()
    result, before = 0, 0
    for s in ox:
        if s == 'O':
            before += 1
            result += before
        else:
            before = 0
    print(result)
```
