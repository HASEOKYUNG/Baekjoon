## 1193번
다음과 같은 배치의 분수표에서 지그재그 방향으로 번호를 부여할 때 N번째 분수를 구하시오.


- 입력) 분수값이 궁금한 N이 주어진다.(1 ≤ N ≤ 10,000,000)
- 출력) N번째 분수를 출력한다.
- https://www.acmicpc.net/problem/1193

### 방법
- 지그재그 순서에 따라 1/1을 시작으로 2개, 3개, 4개, ... 이후 방향이 바뀐다. <br>
  따라서 주어진 N과 가까운 $\sum_{i=1}^k i$의 $k$를 구해 N번째 분수가 $k$번째 행 또는 열에 있음을 찾아야 한다.
- 지그재그 방향은 1/1을 시작으로 오른쪽으로 이동 후 대각선 아래로 내려온 뒤 아래쪽으로 이동 후 대각선 위로 올라가길 반복한다.<br>
  즉, 홀수**행**은 대각선 위로 짝수**열**은 대각선 아래로 이동한다.
- 홀수**행**의 분수는 대각선 위로 움직임에 따라 $N/1$에서 분자는 1씩 줄며 분모는 1씩 늘고 <br>
  짝수**열**의 분수는 대각선 아래로 움직임에 따라 $1/N$에서 분자는 1씩 늘고 분모는 1씩 준다. $k$가 홀수인지, 짝수인지 판별해 규칙을 적용하면 된다.

```python
# 31256 KB, 48 ms
import sys
input = sys.stdin.readline

n = int(input())
k = 0
# N의 최대값에 따라 k의 최댓값은 4472이다.
for i in range(1,4473):
    if n-i > 0:
        n = n-i
        k += 1
    else:
        if k % 2 == 0:
            print(f'{k+2-n}/{n}')
        else:
            print(f'{n}/{k+2-n}')
        break
```

```python
# 31256 KB, 44 ms
import sys
input = sys.stdin.readline

n = int(input())
k = 1
while n - k > 0:
    n -= k
    k += 1
print([f'{n}/{k+1-n}',f'{k+1-n}/{n}'][k%2])
```
