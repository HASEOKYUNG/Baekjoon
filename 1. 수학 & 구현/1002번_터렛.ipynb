## 1002번
중심이 ($x_1, y_1$), ($x_2, y_2$)이고 반지름이 $r_1, r_2$인 두 원의 접점 개수를 구하여라.
- 입력) 첫 줄에 접점 수를 구하는 case 개수가 주어지고 이후 $x_1, y_1, r_1, x_2, y_2, r_2$ 순으로 입력된다.
- 출력) 접점의 개수를 구한다. 단, 접점 수가 무수히 많을 경우 -1을 출력한다.
- https://www.acmicpc.net/problem/1002

### 방법
- 두 원의 중심 사이 거리와 반지름 길이 합, 차이로 아래와 같은 접점 개수를 갖는다.<br>
  (1) 중심 사이 거리가 반지름 길이 합보다 크면 멀리 떨어져 있어 접점이 0개이다.<br>
  (2) 중심 사이 거리가 반지름 길이 합과 같으면 접점은 1개이다.<br>
  (3) 중심 사이 거리가 반지름 길이 합보다 작으면 접점은 2개이다.<br>
  (4) (3)보다 중심 사이 거리가 가까워져 큰 원 안에 작은 원이 내접하면 접점은 1개이다.<br>
  (5) (4)보다 중심 사이 거리가 가까워져 큰 원 안에 작은 원이 속하면 접점은 0개이다.<br>

  ( * ) 두 원이 완전히 겹치면 접점 개수는 무수히 많다.
- 위에서 설명한 규칙에 따라 각 조건에 부합하는 접점의 개수를 출력하면 된다.

```python
# 31256 KB, 60 ms
for i in range(int(input())):
    x1, y1, r1, x2, y2, r2 = map(int, input().split())
    d = ((x1-x2)**2 + (y1-y2)**2)**0.5
    if (d == 0) and (r1==r2):
        print(-1)
    elif (d == r1+r2) or (d == abs(r1-r2)):
        print(1)    
    elif abs(r1-r2) < d < r1+r2:
        print(2)
    else:
        print(0)
```
