## 10798번
최대 15개 글자로 이루어진 문장을 5번 입력받은 뒤 5줄로 나열했을 때, 첫 번째 열부터 마지막 열까지 세로 읽기하라.
- 입력) 빈칸 없이 연속된 문장 5개가 입력된다. 이때, 각 문장의 길이가 다를 수 있다.

- 출력) 세로 읽기를 수행하며<br>
  $~~~~~~~$ 문장 길이가 달라 특정 문장의 읽어야 할 자리에 글자가 없을 경우에는 다음 문장의 글자를 계속 읽는다.

- https://www.acmicpc.net/problem/10798

### 방법1
- 5개 문장을 입력받은 뒤 result 변수에 각 문장의 0부터 가장 긴 문장의 마지막 index값 원소를 합친다.
- 특정 위치에 문자열이 없을 경우 `예외처리`를 통해 pass해 오류가 일어나지 않게 한다.

```python
# 31388 KB, 44 ms
strings = [input() for i in range(5)]

result = ''
for x in range(max(map(len, strings))):
    for y in strings:
        try:
            result += y[x]
        except:
            pass
print(result)
```

### 방법2
- 5개 문장을 입력받고 각 글자를 원소로 하는 행렬을 만든다.
- 각 행의 원소수가 가장 긴 문장의 원소수와 같게 ''를 원소로 추가한다.
- 이후 `zip`으로 세로로 글자를 묶고 join을 사용해 문장을 출력한다.

```python
# 31256 KB, 44 ms
strings = [list(input()) for i in range(5)]

result = ''
m = max(map(len, strings))
for s in zip(*[i+['']*(m-len(i)) for i in strings]):
    result += ''.join(s)
print(result)
```
