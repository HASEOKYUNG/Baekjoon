## 2563번_색종이
가로, 세로의 크기가 각각 100인 정사각형 도화지 위에 최대 100장의 10 x 10 크기의 정사각형 검은 색종이를 붙일 때, 도화지의 검은 색종이로 가려진 넓이 합을 구하라.<br>



- 입력)첫 줄에 색종이 개수가 주어지고 이후 색종이의 좌측하단 좌표가 주어진다.

- 출력)검은 색종이로 가려진 넓이 합을 출력한다.

- https://www.acmicpc.net/problem/2563

### 방법1
- 10 x 10 정사각형 색종이는 1 x 1 정사각형이 100개 모인 것이라 할 수 있으며 1 x 1 정사각형은 1 x 1 정사각형의 꼭짓점 중 하나로 대변할 수 있다.
- 입력받은 정사각형의 100개 1 x 1 정사각형 좌측 하단 꼭짓점 좌표를 구하고 set으로 담는다.<br>
  `적어도` 1개의 색종이에 의해 가려진 좌표만이 set 원소이기에 set의 원소수는 구하고자 하는 넓이의 합과 같다.

```python
# 31256 KB, 44 ms
s = set()

for i in range(int(input())):
    a, b = map(int, input().split())
    s.update({(x, y) for x in range(a, a+10) for y in range(b, b+10)})
print(len(s))
```

### 방법2
- 방법1과 같이 색종이 내 1 x 1 정사각형 좌표 개수를 세는 방법이다.
- 흰 도화지의 모든 좌표를 담을 수 있는 100 x 100 행렬을 만든다.<br>
  이후 색종이 정보를 입력받을 때마다 색종이의 1 x 1 정사각형 좌표 `위치(index)`의 값을 1로 바꾼다.

```python
# 31256 KB, 44 ms
lst = [0]*100*100

for i in range(int(input())):
    a, b = map(int, input().split())
    for x in range(a, a+10):
        for y in range(b, b+10):
            lst[100*x+y] = 1
print(sum(lst))
```
