## 9461번
변의 길이가 1인 정삼각형을 시작으로 나선에서 가장 긴 변에 같은 길이를 갖는 정삼각형을 추가할 때 N번째 추가되는 정삼각형 변의 길이를 구하시오.

- 입력)첫 줄에 테스트 케이스 수가 주어지고 줄단위로 N이 주어진다.
- 출력)각 테스트 케이스에 맞는 N번째 파도반 수열값을 출력한다.
- https://www.acmicpc.net/problem/9461

### 방법
- 도형의 평행 성질에 의해 N회차(단, N > 3) 정삼각형 변의 길이는 N-3회차와 N-2회차 정삼각형 변의 길이를 합해 구할 수 있다.
```python
N = 4, 2 2 _ '4'(1+3)
N = 5,   2 3 _ '5'(1+4)
N = 6,     3 4 _ '7'(2+5)
...
```
- 재귀함수로 해결하고자 하면 n이 클 때 함수 호출 횟수가 늘어나 시간초과 문제가 생긴다. 즉, 다이나믹 프로그래밍으로 해결해야 하는 문제이다.

````python
# 재귀함수
def P(n):
    if n in [1,2,3]:
        return 1
    else:
        return P(n-2)+P(n-3)

for i in range(int(input())):
    print(P(int(input())))
````

````python
# 30616 KB, 44 ms
def P(n):
    p = {1:1,2:1,3:1}
    if n > 3:
        for i in range(4, n+1):
            p[i] = p[i-2]+p[i-3]
    return p[n]

for i in range(int(input())):
    print(P(int(input())))
````
