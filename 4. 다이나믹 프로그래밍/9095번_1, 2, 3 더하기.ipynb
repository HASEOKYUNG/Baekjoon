## 9095번
N을 1, 2, 3의 합으로 표현할 수 있는 수식 개수를 구하시오.
- 입력)첫 줄에 테스트 케이스 수가 주어지고 줄별로 N이 주어진다.
<br>
- 출력)테스트 케이스별 수식 개수를 출력한다.

- https://www.acmicpc.net/problem/9095

### 방법
- 가장 큰 수 3으로 나눈 정수몫에서부터 나머지를 1과 2로 표현하면 된다.<br>
  나머지를 1과 2로 표현하는 방법은 나머지를 2로 나눈 정수몫에서 나머지는 다 1로 표현하면 된다.
- 3으로 나눈 정수몫에서 0까지 1, 2, 3으로 표현한 뒤 배열을 고려해 factorial로 개수를 구한다.

```python
from math import factorial

for i in range(int(input())):
    n = int(input())
    t = 0
    for x in range(n//3, -1, -1):
        for y in range((n-3*x)//2, -1, -1):
            one = n-3*x-2*y
            t += factorial(one+x+y)//(factorial(one)*factorial(x)*factorial(y))
    print(t)
```
