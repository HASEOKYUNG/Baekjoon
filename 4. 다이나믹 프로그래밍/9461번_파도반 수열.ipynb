````python
# 
def P(n):
    if n in [1,2,3]:
        return 1
    else:
        return P(n-2)+P(n-3)

for i in range(int(input())):
    print(P(int(input())))
````

````python
# 30616 KB, 44 ms
def P(n):
    p = {1:1,2:1,3:1}
    if n > 3:
        for i in range(4, n+1):
            p[i] = p[i-2]+p[i-3]
    return p[n]

for i in range(int(input())):
    print(P(int(input())))
````
