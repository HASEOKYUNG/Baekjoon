## 2018번
N을 연속된 자연수의 합으로 표현하는 가짓수를 구하시오.
- 입력)N이 주어진다.($1 \le N \le 10000000$)
  
- 출력)연속된 자연수의 합으로 표현한 가짓수를 출력한다.

<center>(15는 1+2+3+4+5 = 4+5+6 = 7+8 = 15로 표현할 수 있다.)</center>
- https://www.acmicpc.net/problem/2018

### 방법
- range(1, N)에서 start값부터 end값까지의 구간 합을 구한 뒤<br>
  구간 합이 N보다 크면 start값을 하나 키워줘야 해 구간 합에서 현재 start를 빼고 start를 1 더해 기존의 start+1값부터 end값까지 구간 합을 저장시킨다.<br>
  반대로 구간 합이 N보다 작으면 end값을 하나 키워줘야 해 end를 1 더하고 구간 합에 갱신한 end를 더해 start값부터 기존 end+1값까지 구간 합을 저장시킨다.
- 구간 합과 N이 같을 땐 출력해야하는 가짓수로 하나를 추가하고 더 큰 범위를 탐색하기 위해 end값을 하나 키워준다.
- 연속된 자연수의 합으로 start값은 N/2보다 항상 작아야 한다. 따라서 range(1, N) 모두를 탐색할 필요없다.

```python
# 31256 KB, 2972 ms
import sys
input = sys.stdin.readline
n = int(input())

c = 1
s, e, ssum = 1, 2, 1+2
while s < n/2:
    if ssum == n:
        c += 1
        e += 1
        ssum += e
    elif ssum > n:
        ssum -= s
        s += 1
    else:
        e += 1
        ssum += e
print(c)
```
